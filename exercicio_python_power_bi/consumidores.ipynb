Como é possível ver, esse dataframe há 2240 linhas e 29 colunas. Mas para esse exercício não serão usadas todas as colunas, como no caso da variável "ID" que identifica o consumidor. Além das colunas relacionadas à promoção ('AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5' e 'Response'). E também as colunas 'Z_CostContact' e 'Z_Revenue'. 

```python
df.drop('ID', axis=1, inplace= True)
df.drop('AcceptedCmp1', axis=1, inplace= True)
df.drop('AcceptedCmp2', axis=1, inplace= True)
df.drop('AcceptedCmp3', axis=1, inplace= True)
df.drop('AcceptedCmp4', axis=1, inplace= True)
df.drop('AcceptedCmp5', axis=1, inplace= True)
df.drop('Response', axis=1, inplace= True)
df.drop('Z_CostContact', axis=1, inplace= True)
df.drop('Z_Revenue', axis=1, inplace= True)
```

Um outro ponto é que, neste dataframe, apenas a coluna de renda (Income) tem dados nulos, que representam apenas uma pequena parte do total de dados. Por isso iremos tratar esses dados ausentes excluindo as linhas onde se encontra a ausência.

```python
df.dropna(subset=['Income'], inplace=True)
```

Agora, para saber qual faixa etária mais gasta em consumo, será feita uma transformação nos dados. Criando uma coluna nova (Age) para saber a idade dos consumidores a partir do ano de nascimento (Year_Birth).

```python
df['Age'] = 2023 - df['Year_Birth']
```

E uma coluna nova para saber o total gasto considerando todas as categorias de produtos ('MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds')

```python
df['Total_Expense'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

Agora para saber a faixa etária, é preciso agrupar esses consumidores em grupos equivalentes a sua idade.

```python
classes = [0, 40, 50, 60, 70, 80, 140]
labels = ['Ate 40', '41 a 50', '51 a 60', '61 a 70', '71 a 80', '81+']
classes = pd.cut(x=df.Age, bins=classes, labels=labels)

df['Age_Group'] = classes
```

Também é preciso saber o número total de vendas considerando todos os tipos as opções de lugares que os consumidores podem comprar, seja por catálogo (NumCatalogPurchases), no site (NumWebPurchases) ou através de loja física (NumStorePurchases). Para isso será criada uma nova feature contendo todas essas outras juntas.

```python
df['Total_Sales'] = df['NumCatalogPurchases'] + df['NumWebPurchases'] + df['NumStorePurchases']
```
