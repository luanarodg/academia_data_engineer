Como Ã© possÃ­vel ver, esse dataframe hÃ¡ 2240 linhas e 29 colunas. Mas para esse exercÃ­cio nÃ£o serÃ£o usadas todas as colunas, como no caso da variÃ¡vel "ID" que identifica o consumidor. AlÃ©m das colunas relacionadas Ã  promoÃ§Ã£o ('AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5' e 'Response'). E tambÃ©m as colunas 'Z_CostContact' e 'Z_Revenue'. 

```python
df.drop('ID', axis=1, inplace= True)
df.drop('AcceptedCmp1', axis=1, inplace= True)
df.drop('AcceptedCmp2', axis=1, inplace= True)
df.drop('AcceptedCmp3', axis=1, inplace= True)
df.drop('AcceptedCmp4', axis=1, inplace= True)
df.drop('AcceptedCmp5', axis=1, inplace= True)
df.drop('Response', axis=1, inplace= True)
df.drop('Z_CostContact', axis=1, inplace= True)
df.drop('Z_Revenue', axis=1, inplace= True)
```

Um outro ponto Ã© que, neste dataframe, apenas a coluna de renda (Income) tem dados nulos, que representam apenas uma pequena parte do total de dados. Por isso iremos tratar esses dados ausentes excluindo as linhas onde se encontra a ausÃªncia.

```python
df.dropna(subset=['Income'], inplace=True)
```

Agora, para saber qual faixa etÃ¡ria mais gasta em consumo, serÃ¡ feita uma transformaÃ§Ã£o nos dados. Criando uma coluna nova (Age) para saber a idade dos consumidores a partir do ano de nascimento (Year_Birth).

```python
df['Age'] = 2023 - df['Year_Birth']
```

E uma coluna nova para saber o total gasto considerando todas as categorias de produtos ('MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds')

```python
df['Total_Expense'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

Agora para saber a faixa etÃ¡ria, Ã© preciso agrupar esses consumidores em grupos equivalentes a sua idade.

```python
classes = [0, 40, 50, 60, 70, 80, 140]
labels = ['Ate 40', '41 a 50', '51 a 60', '61 a 70', '71 a 80', '81+']
classes = pd.cut(x=df.Age, bins=classes, labels=labels)

df['Age_Group'] = classes
```

TambÃ©m Ã© preciso saber o nÃºmero total de vendas considerando todos os tipos as opÃ§Ãµes de lugares que os consumidores podem comprar, seja por catÃ¡logo (NumCatalogPurchases), no site (NumWebPurchases) ou atravÃ©s de loja fÃ­sica (NumStorePurchases). Para isso serÃ¡ criada uma nova feature contendo todas essas outras juntas.

```python
df['Total_Sales'] = df['NumCatalogPurchases'] + df['NumWebPurchases'] + df['NumStorePurchases']
```
